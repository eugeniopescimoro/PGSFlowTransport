## Localised injection
The source of the solute injected in the simulation domain can take in principle infinite shapes. However, the source geometries which are more relavant to applied studies (contaminant transport, carbon capture and storage, etc.) can be divided into two wide families: local or diffused. The results of the transport simulations performed in this section show the BTCs from solute local sources approximated using a small cubic portion of the domain as a source injection.

## Pluri Gaussian truncated random field using field data: Herten
Herten is a 1120 m<sup>3</sup> depositional formation in the upper Rhine valley of southern Germany. The 16x10x7 meters subsurface parallelepiped was excavated in 1999 and the spatially referenced analogs were released in 2006 along with a detailed description of the hydrogeological properties of its sediments. We then used the permeability values from laboratory tests as reference to populate the domain for our transport experiment.
